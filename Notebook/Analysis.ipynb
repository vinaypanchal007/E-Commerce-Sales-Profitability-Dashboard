```python
conn.execute("DROP VIEW IF EXISTS Monthly_Sales")
query = """
CREATE VIEW Monthly_Sales AS
SELECT
    date(
        substr(l.Order_Date, 7, 4) || '-' ||
        substr(l.Order_Date, 4, 2) || '-' ||
        '01'
    ) AS Month,

    SUM(o.Amount) AS Total_Sales

FROM List_of_Orders AS l
JOIN Order_Details AS o
ON l.Order_ID = o.Order_ID

GROUP BY Month
ORDER BY Month;
"""
conn.execute(query)
print(pd.read_sql_query("SELECT * FROM Monthly_Sales", conn))
```

```python
conn.execute("DROP VIEW IF EXISTS Quarterly_Sales")
query = """
CREATE VIEW Quarterly_Sales AS
SELECT
    substr(l.Order_Date, 7, 4) || '-Q' ||
    CASE
        WHEN CAST(substr(l.Order_Date, 4, 2) AS INTEGER) BETWEEN 1 AND 3 THEN '1'
        WHEN CAST(substr(l.Order_Date, 4, 2) AS INTEGER) BETWEEN 4 AND 6 THEN '2'
        WHEN CAST(substr(l.Order_Date, 4, 2) AS INTEGER) BETWEEN 7 AND 9 THEN '3'
        ELSE '4'
    END AS Quarter,

    SUM(o.Amount) AS Total_Sales

FROM List_of_Orders AS l
JOIN Order_Details AS o
    ON l.Order_ID = o.Order_ID

GROUP BY Quarter
ORDER BY Quarter
"""
conn.execute(query)
result = pd.read_sql_query("SELECT * FROM Quarterly_Sales", conn)
print(result)
```

```python
conn.execute("DROP VIEW IF EXISTS monthly_actual_sales")
query = """
CREATE VIEW monthly_actual_sales AS
SELECT
    (CASE CAST(substr(l.Order_Date,4,2) AS INTEGER)
        WHEN 1 THEN 'Jan' WHEN 2 THEN 'Feb' WHEN 3 THEN 'Mar' WHEN 4 THEN 'Apr'
        WHEN 5 THEN 'May' WHEN 6 THEN 'Jun' WHEN 7 THEN 'Jul' WHEN 8 THEN 'Aug'
        WHEN 9 THEN 'Sep' WHEN 10 THEN 'Oct' WHEN 11 THEN 'Nov' WHEN 12 THEN 'Dec' END)
    || '-' || substr(l.Order_Date,9,2) AS Month_of_Order_Date,
    od.Category,
    SUM(od.Amount) AS Actual_Sales
FROM List_of_Orders l
JOIN Order_Details od ON l.Order_ID = od.Order_ID
GROUP BY 1, 2
"""
conn.execute(query)
print(pd.read_sql_query("SELECT * FROM monthly_actual_sales", conn))
```

```python
conn.execute("DROP VIEW IF EXISTS category_sales_target_vs_actual_monthly")
query2 = """
CREATE VIEW category_sales_target_vs_actual_monthly AS
SELECT
    st.Month_of_Order_Date,
    st.Category,
    st.Target AS Target_Sales,
    COALESCE(ma.Actual_Sales, 0) AS Actual_Sales,
    COALESCE(ma.Actual_Sales, 0) - st.Target AS Variance,
    ROUND(COALESCE(ma.Actual_Sales, 0) * 100.0 / st.Target, 1) AS Target_Achieved_Pct
FROM Sales_Target st
LEFT JOIN monthly_actual_sales ma
    ON st.Category = ma.Category
    AND st.Month_of_Order_Date = ma.Month_of_Order_Date
ORDER BY st.Category, st.Month_of_Order_Date
"""
conn.execute(query2)
result = pd.read_sql_query("SELECT * FROM category_sales_target_vs_actual_monthly", conn)
print(result)
```

```python
conn.execute("DROP VIEW IF EXISTS monthly_sales_target_vs_actual")

query = """
CREATE VIEW monthly_sales_target_vs_actual AS
SELECT
    Month_of_Order_Date,
    SUM(Target_Sales) AS Target_Sales,
    SUM(Actual_Sales) AS Actual_Sales,
    SUM(Variance) AS Variance,
    ROUND(SUM(Actual_Sales) * 100.0 / SUM(Target_Sales), 1) AS Target_Achieved_Pct
FROM category_sales_target_vs_actual_monthly
GROUP BY Month_of_Order_Date
ORDER BY Month_of_Order_Date
"""

conn.execute(query)
print(pd.read_sql_query("SELECT * FROM monthly_sales_target_vs_actual", conn))
```
